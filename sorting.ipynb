```python
# Sorting Algorithms Implementation

# 1. Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


# 2. Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr


# 3. Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr


# 4. Merge Sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


# 5. Quick Sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


# 6. Heap Sort
def heap_sort(arr):
    import heapq
    heapq.heapify(arr)
    return [heapq.heappop(arr) for _ in range(len(arr))]


# -------------------------
# Testing all algorithms
# -------------------------
if __name__ == "__main__":
    data = [64, 34, 25, 12, 22, 11, 90]

    print("Original:", data)

    print("Bubble Sort:", bubble_sort(data.copy()))
    print("Selection Sort:", selection_sort(data.copy()))
    print("Insertion Sort:", insertion_sort(data.copy()))
    print("Merge Sort:", merge_sort(data.copy()))
    print("Quick Sort:", quick_sort(data.copy()))
    print("Heap Sort:", heap_sort(data.copy()))
```

```python
import time
import random
import matplotlib.pyplot as plt
import sys

# ----------------------------
# Sorting Algorithms
# ----------------------------

def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result


def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + mid + quick_sort(right)


# ----------------------------
# Measure Time & Space
# ----------------------------

def measure(sort_func, data):
    arr = data.copy()

    # Space before
    before = sys.getsizeof(arr)

    start = time.time()
    result = sort_func(arr)
    end = time.time()

    # Space after
    after = sys.getsizeof(result)

    time_taken = end - start
    space_used = after - before

    return time_taken, space_used


# ----------------------------
# Input Sizes
# ----------------------------

sizes = [10, 50, 100, 200, 300]

bubble_time = []
merge_time = []
quick_time = []

bubble_space = []
merge_space = []
quick_space = []

for size in sizes:
    data = [random.randint(1, 1000) for _ in range(size)]

    t, s = measure(bubble_sort, data)
    bubble_time.append(t)
    bubble_space.append(s)

    t, s = measure(merge_sort, data)
    merge_time.append(t)
    merge_space.append(s)

    t, s = measure(quick_sort, data)
    quick_time.append(t)
    quick_space.append(s)


# ----------------------------
# Plot Time Graph
# ----------------------------

plt.figure()
plt.plot(sizes, bubble_time, label="Bubble Sort")
plt.plot(sizes, merge_time, label="Merge Sort")
plt.plot(sizes, quick_time, label="Quick Sort")

plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Time Complexity (Time vs Input Size)")
plt.legend()
plt.show()


# ----------------------------
# Plot Space Graph
# ----------------------------

plt.figure()
plt.plot(sizes, bubble_space, label="Bubble Sort")
plt.plot(sizes, merge_space, label="Merge Sort")
plt.plot(sizes, quick_space, label="Quick Sort")

plt.xlabel("Input Size (n)")
plt.ylabel("Memory Used (bytes)")
plt.title("Space Usage (Approximate)")
plt.legend()
plt.show()
```
